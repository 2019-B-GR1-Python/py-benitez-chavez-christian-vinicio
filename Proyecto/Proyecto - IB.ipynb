El siguiente dataset muestra datos de suicidios registrados en distitntos países desde el año de 1987 hasta 2016. Se especificará distintos temas para realziar un análisis completo sobre los suicididos alrededor del mundo.

**2. DESCRIPCIÓN DE COLUMNAS**

| Atributo           | Descripción                                  | 
| ------------------ | :------------------------------------------: | 
| Country            | País donde ocurrió el suicidio               |

| Generation         | Generaciones                                 | 


**3. PROBLEMAS PLANTEADOS**

* **Total de suicidios por país.**

```python
suicide_df.groupby('country').count()["suicides_no"].plot(kind='bar',stacked=True, title="Suicidios por países")
```

* **Total de suicidios en Ecuador por año.**

```python
serie_ecuador = suicide_df.iloc[7694:8065]
```

```python
suicide_df.groupby('year').count()["suicides_no"].plot(kind='bar',stacked=True,title="Total de suicidios en Ecuador por año")
```

* **Cantidad de suicidios totales alrededor de todo el mundo por año.**

```python
suicidios_totales = suicide_df.groupby('year')['suicides_no'].sum()
```

```python
suicidios_totales.plot(kind='line')
```

* **Cantidad total de suicidios entre hombres y mujeres**

```python
anio_sexo = suicide_df.groupby(['year','sex'])
total = anio_sexo['suicides_no'].sum()
suicide_sex = total.unstack()
s_plot = suicide_sex.plot(kind='bar',stacked=True,title="Cantidad total de suicidios entre hombres y mujeres")
s_plot.set_ylabel("Cantidad")
s_plot.set_xlabel("Year")
```

* **Total de suicidios por generacios por mujeres vs hombres**

```python
generacion_sexo = serie_ecuador.groupby(['generation','sex'])
total = generacion_sexo['suicides_no'].sum()
suicidio_por_sexo = total.unstack()
s_plot = suicidio_por_sexo.plot(kind='bar',stacked=True,title="Total de suicidios por generacios por mujeres vs hombres")
s_plot.set_ylabel("Cantidad")
s_plot.set_xlabel("Generation")
```
